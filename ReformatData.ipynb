# Reorganize the given data file

```python
import numpy as np
import pandas as pd
```

Date dictionary to group time series data into one column

```python
# Create date dictionary
d_months = {"2015-01-28":0,
            "2015-02-28":1,
            "2015-03-28":2,
            "2015-04-28":3,
            "2015-05-28":4,
            "2015-06-28":5,
            "2015-07-28":6,
            "2015-08-28":7,
            "2015-09-28":8,
            "2015-10-28":9,
            "2015-11-28":10,
            "2015-12-28":11,
            "2016-01-28":12,
            "2016-02-28":13,
            "2016-03-28":14,
            "2016-04-28":15,
            "2016-05-28":16,
           }

def get_month(s):
    return d_months[s]

# Create customer dictionary
customer_dict = dict()
```

Customer class to handle data collection. Will be later on exported to text file.

```python
class Customer:
    def __init__(self):
        self.customer_code = ""
        self.employee_index = [""] * 17
        self.country = ""
        self.sex = ""
        self.age = ""
        self.date_begin_contract = ""
        self.new_customer_index = [""] * 17
        self.seniority = [""] * 17
        self.primary_at_month_end = [""] * 17
        self.last_date_as_primary = [""] * 17
        self.customer_type_begin_month = [""] * 17
        self.relation_type_begin_month = [""] * 17
        self.residence_index = ""
        self.foreigner_index = ""
        self.spouse_index = ""
        self.join_channel = ""
        self.deceased_index = ""
        self.address_type = ""
        self.province_code = ""
        self.province_name = ""
        self.activity_index = [""] * 17
        self.gross_income = ""
        self.segmentation = [""] * 17
        self.Saving_Account = [""] * 17
        self.Guarantees = [""] * 17
        self.Current_Accounts = [""] * 17
        self.Derivada_Account = [""] * 17
        self.Payroll_Account = [""] * 17
        self.Junior_Account = [""] * 17
        self.Mas_Particular_Account = [""] * 17
        self.Particular_Account = [""] * 17
        self.Particular_Plus_Account = [""] * 17
        self.Short_Term_Deposits = [""] * 17
        self.Medium_Term_Deposits = [""] * 17
        self.Long_Term_Deposite = [""] * 17
        self.E_Account = [""] * 17
        self.Funds = [""] * 17
        self.Mortgage = [""] * 17
        self.Pensions_1 = [""] * 17
        self.Loans = [""] * 17
        self.Taxes = [""] * 17
        self.Credit_Card = [""] * 17
        self.Securities = [""] * 17
        self.Home_Account = [""] * 17
        self.Payroll = [""] * 17
        self.Pensions_2 = [""] * 17
        self.Direct_Debit = [""] * 17
        
    def __str__(self):
        return ','.join([
                self.customer_code, 
                '|'.join(self.employee_index), 
                self.country, 
                self.sex, 
                self.age, 
                self.date_begin_contract, 
                '|'.join(self.new_customer_index), 
                '|'.join(self.seniority), 
                '|'.join(self.primary_at_month_end), 
                '|'.join(self.last_date_as_primary), 
                '|'.join(self.customer_type_begin_month), 
                '|'.join(self.relation_type_begin_month), 
                self.residence_index, 
                self.foreigner_index, 
                self.spouse_index, 
                self.join_channel, 
                self.deceased_index, 
                self.address_type, 
                self.province_code, 
                self.province_name, 
                '|'.join(self.activity_index), 
                self.gross_income, 
                '|'.join(self.segmentation),
                '|'.join(self.Saving_Account), 
                '|'.join(self.Guarantees), 
                '|'.join(self.Current_Accounts), 
                '|'.join(self.Derivada_Account), 
                '|'.join(self.Payroll_Account), 
                '|'.join(self.Junior_Account), 
                '|'.join(self.Mas_Particular_Account), 
                '|'.join(self.Particular_Account), 
                '|'.join(self.Particular_Plus_Account), 
                '|'.join(self.Short_Term_Deposits), 
                '|'.join(self.Medium_Term_Deposits), 
                '|'.join(self.Long_Term_Deposite), 
                '|'.join(self.E_Account), 
                '|'.join(self.Funds), 
                '|'.join(self.Mortgage), 
                '|'.join(self.Pensions_1), 
                '|'.join(self.Loans), 
                '|'.join(self.Taxes), 
                '|'.join(self.Credit_Card), 
                '|'.join(self.Securities), 
                '|'.join(self.Home_Account), 
                '|'.join(self.Payroll),
                '|'.join(self.Pensions_2), 
                '|'.join(self.Direct_Debit)
                ]) + '\n'
```

Import the original file and collect all the necessary pieces into the class structures

```python
# Process the huge file
import csv
with open("Data/train_ver2.csv") as myfile:
    #Initiate csv reader
    reader = csv.reader(myfile, delimiter = ',', quotechar='"')
    #Skip header line
    reader.next()
    
    # Process the data
    for fields in reader:
        id = int(fields[1])
        
        # Detect the customer
        c = customer_dict.get(id, None)
        if (c == None):
            c = Customer()
            customer_dict[id] = c
        
        # Edit the customer
        
        # Process data that changes over time
        m = get_month(fields[0])
        
        c.employee_index[m] = fields[2].strip()
        c.new_customer_index[m] = fields[7].strip()
        c.seniority[m] = fields[8].strip()
        c.primary_at_month_end[m] = fields[9].strip()
        c.last_date_as_primary[m] = fields[10].strip()
        c.customer_type_begin_month[m] = fields[11].strip()
        c.relation_type_begin_month[m] = fields[12].strip()
        c.activity_index[m] = fields[21].strip()
        c.segmentation[m] = "" if (fields[23] == "") else (fields[23].split()[0])
        c.Saving_Account[m] = fields[24].strip()
        c.Guarantees[m] = fields[25].strip()
        c.Current_Accounts[m] = fields[26].strip()
        c.Derivada_Account[m] = fields[27].strip()
        c.Payroll_Account[m] = fields[28].strip()
        c.Junior_Account[m] = fields[29].strip()
        c.Mas_Particular_Account[m] = fields[30].strip()
        c.Particular_Account[m] = fields[31].strip()
        c.Particular_Plus_Account[m] = fields[32].strip()
        c.Short_Term_Deposits[m] = fields[33].strip()
        c.Medium_Term_Deposits[m] = fields[34].strip()
        c.Long_Term_Deposite[m] = fields[35].strip()
        c.E_Account[m] = fields[36].strip()
        c.Funds[m] = fields[37].strip()
        c.Mortgage[m] = fields[38].strip()
        c.Pensions_1[m] = fields[39].strip()
        c.Loans[m] = fields[40].strip()
        c.Taxes[m] = fields[41].strip()
        c.Credit_Card[m] = fields[42].strip()
        c.Securities[m] = fields[43].strip()
        c.Home_Account[m] = fields[44].strip()
        c.Payroll[m] = fields[45].strip()
        c.Pensions_2[m] = fields[46].strip()
        c.Direct_Debit[m] = fields[47].strip()
        
        # Process data that stays constant or change likely has no implication
        customer_code_temp = fields[1].strip()
        country_temp = fields[3].strip()
        sex_temp = fields[4].strip()
        age_temp = fields[5].strip()
        date_begin_contract_temp = fields[6].strip()
        residence_index_temp = fields[13].strip()
        foreigner_index_temp = fields[14].strip()
        spouse_index_temp = fields[15].strip()
        join_channel_temp = fields[16].strip()
        deceased_index_temp = fields[17].strip()
        address_type_temp = fields[18].strip()
        province_code_temp = fields[19].strip()
        province_name_temp = '"' + fields[20].strip() + '"'
        gross_income_temp = fields[22].strip()
        
        
        if customer_code_temp != "":
            c.customer_code = customer_code_temp
        if country_temp != "":
            c.country = country_temp
        if sex_temp != "":
            c.sex = sex_temp
        if age_temp != "":
            c.age = age_temp
        if date_begin_contract_temp != "":
            c.date_begin_contract = date_begin_contract_temp
        if residence_index_temp != "":
            c.residence_index = residence_index_temp
        if foreigner_index_temp != "":
            c.foreigner_index = foreigner_index_temp
        if spouse_index_temp != "":
            c.spouse_index = spouse_index_temp
        if join_channel_temp != "":
            c.join_channel = join_channel_temp
        if deceased_index_temp != "":
            c.deceased_index = deceased_index_temp
        if address_type_temp != "":
            c.address_type = address_type_temp
        if province_code_temp != "":
            c.province_code = province_code_temp
        if province_name_temp != "":
            c.province_name = province_name_temp
        if gross_income_temp != "":
            c.gross_income = gross_income_temp
```
